# Axes Labels,Legend and Grid

# Bar chart

```python
product = [1,2,3,4]
price = [15,25,40,30]
plt.bar(product,price,label="Price")
```

```python
product = [1,2,3,4]
price = [15,25,40,30]
plt.bar(product,price,label="Price",width=.5)
```

```python
product = [1,2,3,4]
price = [15,25,40,30]
plt.bar(product,price,label="Price",width=.5,color='green')
```

```python
product = [1,2,3,4]
price = [15,25,40,30]
profit = [5,7,10,5]
plt.bar(product,price,label="Price",width=.5,color='green')
plt.bar(product,profit,label="profit")
```

```python
product = [1,2,3,4]
price = [15,25,40,30]
profit = [5,7,10,5]
plt.bar(product,price,label="Price",color='green')
plt.bar(product,profit,label="profit")
```

```python
import numpy as np
product = np.array([1,2,3,4])
price = [15,25,40,30]
profit = [5,7,10,5]
plt.bar(product-.2,price,label="Price",color='green',width=.4)
plt.bar(product+.2,profit,label="profit",color="red",width=.4)
```

```python
import numpy as np
product = np.array([1,2,3,4])
price = [15,25,40,30]
profit = [5,7,10,5]
plt.bar(product-.2,price,label="Price",color='green',width=.4)
plt.bar(product+.3,profit,label="profit",color="red",width=.4)
```

# pie chart

```python
import matplotlib.pyplot as plt
price = [10,5,10,10,5,40]
lab = ['khata','pen','pencil','eraser','taka','total']
plt.pie(price)
```

```python
plt.pie(price,labels=lab)
```

```python
plt.pie(price,labels=lab,radius=2)
```

```python
plt.pie(price,labels=lab,radius=2,autopct='%.2f%%' )
```

```python
plt.pie(price,labels=lab,radius=2,autopct='%.2f%%',shadow=True )
```

```python
plt.pie(price,labels=lab,radius=2,autopct='%.2f%%',shadow=True,startangle=45)
```

```python
plt.pie(price,labels=lab,radius=2,autopct='%.2f%%',shadow=True,startangle=45,explode=[0,0,0,0,0,0])
```

```python
plt.pie(price,labels=lab,radius=2,autopct='%.2f%%',shadow=True,startangle=45,explode=[.2,0,0,0,0,0])
```
